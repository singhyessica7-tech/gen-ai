variable length argument: we give any number of argument

1. **non keyword variable length argument**: a.it behaves like a tuple.

b.it is represented by a star(*)


2.**keyword variable length argument**:a.it behaves like a dictionary

b. it is represented by double star(**)

**non keyword variable length argument**::

```python
def test(*args):
    print(args)
    print(len(args))
    print(type(args))

```

```python
test(1,2,3,4,5,6)
```

```python
def sum_number(*args):
    sum=0
    for num in args:
      sum+=num
    return sum
```

```python
sum_number(1,2,3,4,5,6,7,8,9,55,77,44,66,33,22)
```

```python
9def test(**kwargs):
  print(kwargs)
  print(len(kwargs))
  print(type(kwargs))
```

```python
#bar plot : it works on numerical as well as categorical data
import matplotlib.pyplot as plt
import numpy as np
fruit=np.array(["Apple","Banana","Orange","Watermelon"])
quantity=np.array([100,120,80,70])
color=np.array(["red","yellow","orange","green"])
plt.bar(fruit,quantity,color=color)
plt.xlabel=("Fruit")
plt.ylabel("Quantity of fruit")
plt.title("Fruit vs fruit quantity")
plt.show()


```

```python
fruit=np.array(["Apple","Banana","Orange","Watermelon"])
quantity=np.array([100,120,80,70])
color=np.array(["red","yellow","orange","green"])
plt.barh(fruit,quantity,color = color)
plt.xlabel=("Fruit")
plt.ylabel("Quantity of fruit")
plt.title("Fruit vs fruit quantity")
plt.show()
```

```python
#pie plot
percentage_of_fruit=np.array([20,25,30,25])
fruit=np.array(["Apple","Banana","Orange","Watermelon"])
myexplode=np.array([0,0,0,0.2])
plt.pie(percentage_of_fruit,labels=fruit, explode=myexplode)
plt.legend(title="Fruit",loc="best")
plt.show()
```

```python
import numpy as np
from numpy.linalg import norm

A = np.array([2,1,2,3,2,9])
B = np.array([3,4,2,4,5,5])

#compute cosine similarity
cosine = np.dot(A,B) / (norm(A) * norm(B))
print("Cosine Similarity:" , cosine)
```

```python
from transformers import AutoTokenizer

#load a tokenizer(e.g., BERT)
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text= "Machine learning is transforming materials science!"

#tokenize
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text)

print("tokens:",tokens)
print("token IDS:", token_ids)
```
